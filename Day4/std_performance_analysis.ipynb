```python
import pandas as pd 
```

```python
df = pd.read_csv("student_performance.csv")
```

```python
df.info()
```

```python
df.columns
```

```python
subject_marks = df.drop(columns=['Student_ID', 'Name', 'Age', 'Program', 'Attendance'])
```

```python
subject_marks
```

```python
average_marks = subject_marks.mean().to_dict()
average_marks
```

```python
df['Total Marks'] = subject_marks.sum(axis=1)
subject_marks['Total Marks'] = subject_marks.sum(axis=1)
```

```python
subject_marks
```

```python
df
```

```python
top_five_students = df.sort_values("Total Marks", ascending=False).head()
```

```python
top_five_students
```

```python
average_marks
```

```python
for subject in average_marks:
    low_scorers = (subject_marks.loc[subject_marks[subject] < average_marks[subject]].index.tolist())
    print("Students scoring below average in", subject, "are:\n", df.loc[low_scorers,['Student_ID','Name', subject]], "\n")
```

```python
df['Total Marks'].mean()
```

```python
print("Students scoring below average overall:")
low_scorers = df.loc[df['Total Marks'] < df['Total Marks'].mean(), ['Student_ID','Name','Total Marks']]
low_scorers
```

```python
print(len(df))  # total no of students
```

```python
low_scorers.to_csv('low_scorers.csv', index=False)
top_five_students.to_csv('top_five_students.csv', index=False)
df.to_csv('record_with_totals.csv', index=False)
```
