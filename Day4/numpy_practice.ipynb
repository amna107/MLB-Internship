# Problem 4 

```python
# Reshaping
a = np.array([[1,2,3,4],[5,6,7,8]])
print(a)
print(a.shape)

a = a.reshape((4,2))
print(a)
print(a.shape)
```

# Problem 5

```python
# slice and get only last two rows
print(a[2:])

# slice and get only second column
print(a[:, 1:])

# slice and index lets say we want print only 6
print(a[2, 1:])

# now we want print only 6 and 7
print(a[2, 1:], a[3, :1])
```

```python
# using indexing to change an element 
a = np.array([[1,2,3,4],[5,6,7,8]])
print(a)

# change 8 to 16
a[-1, 3:] *= 2
print(a)
```

```python
# true matix multiplication (order of the matrix matters)
a = np.ones((2,3))
print(a.shape) 
print(a)
b = np.full((3,2), 2)
print(b.shape)
print(b)
print(np.matmul(a,b))
```
