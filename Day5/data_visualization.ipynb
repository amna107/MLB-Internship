```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
import os
```

```python
# loading the dataset 

df = pd.read_csv("cleaned_student_performance.csv")
```

```python
os.makedirs("Charts/Practice/", exist_ok=False)
```

```python
sns.barplot(data=df, x='Name', y='Average_Score')
plt.xticks(rotation=45)
plt.ylabel("Average marks")
plt.title("Average marks of each student")
plt.savefig("Charts/Practice/barplot.png", bbox_inches='tight')
plt.show()
plt.close()
```

```python
sns.histplot(df, x="Average_Score", kde=True)
plt.title("Distribution of Average Scores")
plt.savefig("Charts/Practice/histplot.png", bbox_inches='tight')
plt.show()
plt.close()
```

```python
sns.scatterplot(df, x="Python", y="ML")
plt.xlabel("Python marks") 
plt.ylabel("ML marks") 
plt.title("Python vs ML marks")
plt.savefig("Charts/Practice/scatterplot.png", bbox_inches='tight')
plt.show()
plt.close()
```

```python
category_count = df['Performance'].value_counts()
print(category_count)

plt.figure(figsize = (10,10))
plt.pie(category_count, labels=category_count.index, autopct="%1.1f%%")
plt.legend(loc="upper right")
plt.savefig("Charts/Practice/pieplot.png", bbox_inches='tight')
plt.show()
plt.close()
```

```python
df.columns
sns.boxplot(df[['Python', 'Mathematics','Statistics', 'ML']])
plt.title("Spread of marks in all subjects")
plt.savefig("Charts/Practice/boxplot.png", bbox_inches='tight')
plt.show()
plt.close()
```
