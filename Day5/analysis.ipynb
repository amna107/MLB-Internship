```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

How many students are in the dataset?

```python
df = pd.read_csv("cleaned_student_performance.csv")
print("The dataset has", len(df), "students.")
```

What is the average score for each subject?

```python
df.columns
marks = df[['Python', 'Mathematics','Statistics', 'ML']]
print("The average scores for each subject are:")
avg = marks.mean()
avg
```

```python
ax = sns.barplot(avg)
ax.bar_label(ax.containers[0]) # conatiners[0] conatins the whole bars list, [0][0] then just python
plt.xlabel("Subjects")
plt.ylabel("Average marks")
plt.title("Average marks comparison")
plt.savefig("Charts/Project/1.png", bbox_inches='tight')
plt.show()
plt.close()
```

Who are the Top 5 performing students?

```python
df['Total Marks'] = marks.sum(axis=1)
top = df.sort_values(by="Total Marks",ascending = False)
```

```python
top_five = top.reset_index(drop=True).head()
```

```python
print("The top five students are:")
print(top_five['Name'])
```

```python
plt.figure(figsize=(20,6))

plt.subplot(1,2,1)
ax = sns.barplot(top, x='Total Marks', y='Name')
ax.bar_label(ax.containers[0])
plt.title("ALL")
plt.ylabel("Students")

plt.subplot(1,2,2)
ax = sns.barplot(top_five, x='Total Marks', y='Name')
ax.bar_label(ax.containers[0])
plt.title("TOP 5")
plt.ylabel("")

plt.savefig("Charts/Project/2.png", bbox_inches='tight')
plt.show()
plt.close()
```

Which students need improvement?

```python
print("Studnets who need improvement are:")
need_improve = df.loc[df['Performance']=="Needs Improvement", 'Name']
print(need_improve.to_string(index=False))  # .to_string() will stop rendering it as a table and hence return us a list without index becuase we exculded that
```

```python
count = df['Performance'].value_counts()

bars = plt.bar(count.index, count.values)
plt.bar_label(bars)
ax = plt.gca()                              # get current axes
ax.set_yticks([])   #removes both label and values form y axis
plt.title("Class Performance")
plt.ylabel("No. of Students")
plt.xlabel("Perforamnce Category")

plt.savefig("Charts/Project/3.png", bbox_inches='tight')
plt.show()
plt.close()
```
